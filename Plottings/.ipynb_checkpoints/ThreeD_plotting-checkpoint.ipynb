```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f(x, y):
    z = x**2 + y**2
    return z
```

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.contour3D(X, Y, Z, 50, cmap=plt.cm.jet)
```
